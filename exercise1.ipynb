Q1. Write a function that takes a string and returns:
-reversed string
-string with every 2nd character

```python
def string_operations(text):
    reversed_text = text[::-1]
    every_second_char = text[1::2]
    return reversed_text, every_second_char

sample = "python"
rev_text, second_chars = string_operations(sample)
print("Reversed:", rev_text)
print("Every 2nd character:", second_chars)
```

Q2. Given a list of numbers, return a new list containing squares of even numbers only using list comprehension.

```python
def square_even_numbers(numbers):
    return [num * num for num in numbers if num % 2 == 0]

nums = [1, 2, 3, 4, 5, 6]
print(square_even_numbers(nums))
```

Q3. Write a function that counts frequency of each character in a string using a dictionary.

```python
def char_frequency(text):
    freq = {}
    for ch in text:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1
    return freq

print(char_frequency("banana"))
```

Q4. Write a function that takes a list and returns indices of all odd numbers using enumerate().

```python
def odd_indices(values):
    indices = []
    for i, value in enumerate(values):
        if value % 2 != 0:
            indices.append(i)
    return indices

print(odd_indices([10, 7, 4, 9, 13, 2]))
```

Q5. Write a function to rotate a list right by k positions using slicing.

```python
def rotate_right(lst, k):
    if len(lst) == 0:
        return []

    k = k % len(lst)
    if k == 0:
        return lst[:]

    return lst[-k:] + lst[:-k]

print(rotate_right([1, 2, 3, 4, 5], 2))
```

Q6. Write a function that takes two lists and returns a dictionary using zip().

```python
def lists_to_dict(keys, values):
    result = dict(zip(keys, values))
    return result

keys = ["name", "age", "city"]
values = ["Ali", 20, "Lahore"]
print(lists_to_dict(keys, values))
```

Q7. Using iter() and next(), manually iterate through a list and print elements one by one. Also, perform exception handling for the exception StopIteration using a try-except block.

```python
def manual_iteration(items):
    iterator_obj = iter(items)

    while True:
        try:
            value = next(iterator_obj)
            print(value)
        except StopIteration:
            print("StopIteration occurred: reached end of list")
            break

manual_iteration([5, 10, 15, 20])
```

Q8. Write a function that flattens a nested list using list comprehension.

```python
def flatten_list(nested_list):
    flat = [item for sublist in nested_list for item in sublist]
    return flat

print(flatten_list([[1, 2], [3, 4], [5, 6]]))
```

Q9. Write a function that checks whether a string is a palindrome using slicing.

```python
def is_palindrome(text):
    cleaned = text.lower().replace(" ", "")
    return cleaned == cleaned[::-1]

print(is_palindrome("madam"))
print(is_palindrome("world"))
```

Q10. Write a function that returns the second largest number in a list (without sorting).

```python
def second_largest(nums):
    if len(nums) < 2:
        return None

    largest = float("-inf")
    second = float("-inf")

    for num in nums:
        if num > largest:
            second = largest
            largest = num
        elif largest > num > second:
            second = num

    if second == float("-inf"):
        return None

    return second

print(second_largest([10, 4, 8, 20, 15]))
print(second_largest([5, 5, 5]))
```
